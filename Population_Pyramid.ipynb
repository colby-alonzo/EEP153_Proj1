```python
# Brazilian male population by age
male_pop_variable_labels = {"SP.POP.0004.MA":"M0-4","SP.POP.0509.MA":"M5-9","SP.POP.1014.MA":"M10-15",
                   "SP.POP.1519.MA":"M15-19","SP.POP.2024.MA":"M20-24",
                  "SP.POP.2529.MA":"M25-29","SP.POP.3034.MA":"M30-34",
                  "SP.POP.3539.MA":"M35-39","SP.POP.4044.MA":"M40-44",
                  "SP.POP.4549.MA":"M45-49","SP.POP.5054.MA":"M50-54",
                  "SP.POP.5559.MA":"M55-59","SP.POP.6064.MA":"M60-64",
                  "SP.POP.6569.MA":"M65-69","SP.POP.7074.MA":"M70-74",
                  "SP.POP.7579.MA":"M75-79","SP.POP.80UP.MA":"M80"}

# Filter country and date
# arbitrarily chose 2019, feel free to change
male_pop_BRA_2019 = wbdata.get_dataframe(male_pop_variable_labels, country="BRA", date="2019-01-01")

# Display first few rows
male_pop_BRA_2019.head()
```

```python
#Brazilian female population by age
female_pop_variable_labels = {"SP.POP.0004.FE":"F0-4","SP.POP.0509.FE":"F5-9","SP.POP.1014.FE":"F10-15",
                   "SP.POP.1519.FE":"F15-19","SP.POP.2024.FE":"F20-24",
                  "SP.POP.2529.FE":"F25-29","SP.POP.3034.FE":"F30-34",
                  "SP.POP.3539.FE":"F35-39","SP.POP.4044.FE":"F40-44",
                  "SP.POP.4549.FE":"F45-49","SP.POP.5054.FE":"F50-54",
                  "SP.POP.5559.FE":"F55-59","SP.POP.6064.FE":"F60-64",
                  "SP.POP.6569.FE":"F65-69","SP.POP.7074.FE":"F70-74",
                  "SP.POP.7579.FE":"F75-79","SP.POP.80UP.FE":"F80"}

# Filter country and date
# arbitrarily chose 2019, feel free to change
female_pop_BRA_2019 = wbdata.get_dataframe(female_pop_variable_labels, country="BRA", date="2019-01-01")

# Display first few rows
female_pop_BRA_2019.head()
```

```python
male_pop_pyramid = male_pop_BRA_2019.to_numpy().flatten()
female_pop_pyramid = female_pop_BRA_2019.to_numpy().flatten()
```

```python
import matplotlib.pyplot as plt
import numpy as np

age_groups = ["0-4", "5-9", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40-44", "45-49", "50-54", "55-59", "60-64", "65-69", "70-74", "75-79", "80+"]

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 8))

# Plot male population
ax.barh(age_groups, male_pop_pyramid, color='teal', label='Male')

# Plot female population
ax.barh(age_groups, [-x for x in female_pop_pyramid], color='orchid', label='Female')

ax.set_title('Brazil Population Pyramid (2019)')
ax.set_xlabel('Population')
ax.set_ylabel('Age Group')
ax.legend()

plt.show()
```
